```python
url = 'https://www.olx.co.id/all-results/q-mobil/'
req_mobil = requests.get(url)
soup_mobil = BeautifulSoup(req_mobil.text, 'lxml')
```

```python
seluruhmobil = soup_mobil.find_all('a',{'class':'marginright5 link linkWithHash detailsLink'})
```

```python
for mobil in seluruhmobil :
        print(mobil['href'])
```

```python
for judul in seluruhmobil :
    print(judul.text)
```

```python
seluruh_harga = soup_mobil.find_all('strong',{'class':'c000'})
```

```python
for harga in seluruh_harga :
    print(harga.text)
```

```python
kota = soup_mobil.find_all('p',{'class':'color-9 lheight14 margintop3'})
kota
```

```python
for kotaku in kota :
    print(kotaku.find('span').text)
```

```python
tanggal = soup_mobil.find_all('td', class_='rel tcenter')
```

```python
for tanggalku in tanggal :
    print(tanggalku.text.strip())
```

```python
gambar = soup_mobil.find_all('img', class_='fleft')
```

```python
for gambarku in gambar :
    print(gambarku['src'])
```

```python
mobil = []
for judul, harga, tanggalku, kotaku, gambarku in zip(seluruhmobil, seluruh_harga, tanggal, kota, gambar) :
    item = {}
    print(judul.text.strip())
    item['judul']=judul.text.strip()
    print(harga.text.strip())
    item['harga'] = harga.text.strip()
    print(tanggalku.text.strip())
    item['tanggal'] = tanggalku.text.strip()
    print(kotaku.find('span').text.strip())
    item['kota'] = kotaku.find('span').text.strip()
    print(gambarku['src'].strip())
    item['gambar'] = gambarku['src'].strip()
    mobil.append(item)
```
